```python
from instagrapi import Client
import json
import datetime
```

```python
# ----- CONFIG load -----
with open("../data/config.json", "r") as f:
    js = json.load(f)

IG = js['instagram']
USERNAME = IG['username']
PASSWORD = IG['password']
TARGET_USERNAME = IG['target_username']  # the username of the person you're searching for in threads
```

```python
cl = Client()
cl.login(USERNAME, PASSWORD)
```

```python
target_user_id = cl.user_id_from_username(TARGET_USERNAME)
```

```python
target_user_id
```

```python
thread = cl.direct_thread_by_participants(user_ids=[target_user_id])
```

```python
thread
```

```python
thread_data = thread['thread']

# The thread ID is found under the 'thread_v2_id' key
thread_id = thread_data['thread_v2_id']

print(f"The Thread ID is: {thread_id}")
```

```python
messages = cl.direct_messages(thread_id, 0)	
```

```python
messages
```

```python
data = []
for message in messages:
    data.append({
        "sender": "Jasmin" if message.user_id == target_user_id else "Austin",
        "timestamp": message.timestamp.timestamp(),
        "item_type": message.item_type,
        "text": message.text,
        "object": str(message),
        # "objectRef": message
    })
```

```python
data
```

```python
with open("../data/messages.json", 'w') as f:
    json.dump(data, f)
```

```python
messages_count = {
    "Austin": 0,
    "Jasmin": 0
}

for d in data:
    messages_count[d['sender']] += 1

messages_count
```

```python
reels_count = {
    "Austin": 0,
    "Jasmin": 0
}

for d in data:
    if (d['item_type'] == ""):
        reels_count[d['sender']] += 1

reels_count
```

```python

frequencies = {
    "Austin": {},
    "Jasmin": {}
}

for message in data:
    key = message['sender']
    if (message['item_type'] != "text"):
        continue

    m = message['text'].split(" ")
    for word in m:
        word = word.lower()
        if (word not in frequencies[key]):
            frequencies[key][word] = 1
        else:
            frequencies[key][word] += 1

frequencies["Austin"] = dict(sorted(frequencies["Austin"].items(), key=lambda x: -1 * x[1]))
frequencies["Jasmin"] = dict(sorted(frequencies["Jasmin"].items(), key=lambda x: -1 * x[1]))
frequencies
```
